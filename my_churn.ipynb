```python
import pandas as pd
import matplotlib.pyplot as plt
churn_data = pd.read_csv('churn.csv',sep=',')
churn_data.drop('RowNumber',axis=1,inplace=True)
churn_data['exited_data'] = churn_data['Exited'].apply(lambda x:'Лояльный' if x ==0 else 'Ушедший')
result = churn_data['exited_data'].value_counts()

fig = plt.figure(figsize=(15,5))
axes= fig.add_axes([0,0,1,1])
axes.bar(x = result.index, height= result)
axes.set_title('Cоотношение ушедших и лояльных клиентов',fontsize = 16)
axes.set_xlabel('Категории клиентов')
axes.set_ylabel('Общее колличество в каждой категории')
axes.xaxis.set_tick_params(rotation = 45)



```
